```python
#Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import json
from datetime import datetime
```

```python
#insert the website link 
#hospitalisation, cases number, death and RAT cases.
table_hos = []
table_cases = []
table_death = []
table_rat = []

states = ["vic", "nsw", "qld", "wa", "sa","tas","act", "nt" ]
for x in states:
    url_hos = (f"https://covidlive.com.au/report/daily-hospitalised/{x}")
    url_cases = (f"https://covidlive.com.au/report/daily-cases/{x}")
    url_deaths = (f"https://covidlive.com.au/report/daily-deaths/{x}") 
    url_rat = (f"https://covidlive.com.au/report/daily-cases-probable/{x}") 
    
    # retrieve database:
    table_hos1 = pd.read_html(url_hos)[1]
    table_cases1 = pd.read_html(url_cases)[1]
    table_death1 = pd.read_html(url_deaths)[1]
    table_rat1 = pd.read_html(url_rat)[1]

    #append to the list:
    table_hos.append(table_hos1)
    table_cases.append(table_cases1)
    table_death.append(table_death1)
    table_rat.append(table_rat1)
    
```

```python
#VIC hospitalisation, Cases number, Death number and RAT cases
vic_hos = table_hos[0].copy()
vic_cases = table_cases[0].copy()
vic_death = table_death[0].copy()
vic_rat = table_rat[0].copy()

#NSW hospitalisation, Cases number, Death number and RAT cases
nsw_hos = table_hos[1].copy()
nsw_cases = table_cases[1].copy()
nsw_death = table_death[1].copy()
nsw_rat = table_rat[1].copy()

#QLD hospitalisation, Cases number, Death number and RAT cases
qld_hos = table_hos[2].copy()
qld_cases = table_cases[2].copy()
qld_death = table_death[2].copy()
qld_rat = table_rat[2].copy()

#WA hospitalisation, Cases number, Death number and RAT cases
wa_hos = table_hos[3].copy()
wa_cases = table_cases[3].copy()
wa_death = table_death[3].copy()
wa_rat = table_rat[3].copy()

#SA hospitalisation, Cases number, Death number and RAT cases
sa_hos = table_hos[4].copy()
sa_cases = table_cases[4].copy()
sa_death = table_death[4].copy()
sa_rat = table_rat[4].copy()

#TAS hospitalisation, Cases number, Death number and RAT cases
tas_hos = table_hos[5].copy()
tas_cases = table_cases[5].copy()
tas_death = table_death[5].copy()
tas_rat = table_rat[5].copy()

#ACT hospitalisation, Cases number, Death number and RAT cases
act_hos = table_hos[6].copy()
act_cases = table_cases[6].copy()
act_death = table_death[6].copy()
act_rat = table_rat[6].copy()

#NT hospitalisation, Cases number, Death number and RAT cases
nt_hos = table_hos[7].copy()
nt_cases = table_cases[7].copy()
nt_death = table_death[7].copy()
nt_rat = table_rat[7].copy()
```
